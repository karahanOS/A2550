```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
a_real = 3
b_real = 4

x = np.linspace(0,100, 100)
y = line(x, a_real, b_real)
y_err = np.random.normal(0,2,len(x))

#print(x)
#print(y)
#print(y_err)

y = y + y_err

#print(y)
```

```python
plt.scatter(x, y, color="black", s=5, label="measurement")

plt.title("Line Fit")

plt.xlabel("X values")
plt.ylabel("Y values")

plt.xlim(left=0)
plt.ylim(bottom=0)

plt.legend()

plt.show()
```

```python
def line(x, a, b):
    return a*x + b

parameters, covariance = curve_fit(line, x, y)
```

```python
a_fit, b_fit = parameters
print(a_fit, b_fit)
```

```python
x_fit = np.linspace(0, 100, 1000)
y_fit = line(x_fit, a_fit, b_fit)


plt.scatter(x, y, color="black", s=5, label="data")
plt.plot(x_fit,y_fit, color="red", label="fit: f(x)={:.2f}.x+{:.2f}".format(a_fit,b_fit))

plt.title("Line Fit")

plt.xlabel("X values")
plt.ylabel("Y values")

plt.xlim(left=0)
plt.ylim(bottom=0)

plt.legend()

plt.show()
```
